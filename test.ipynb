```python
import json

# Open and read the JSON file
with open('data/data_carbon.json', 'r') as file:
    data = json.load(file)
```

```python
elements = []

for child in data['children']:
    for subchild in child['children']:
        for item in subchild['children']:
            elements.append(item['name'])
```

```python
# the json file is hierarchical, with children and parents. Print the 2nd layer


# the file looks like this:

# {
#   "name": "Food",
#   "children": [
#     {
#       "name": "animal products",
#       "value": 9.43,
#       "children": [
#         {
#           "name": "beef and buffallo meat",
#           "value": 40.84,
#           "children": [
#             {
#               "name": "beef bone free meat",
#               "value": 25.75
#             },
#             {
#               "name": "beef meat with bone",
#               "value": 17.96
#             }

# print the names of the children of the first layer

for child in data['children']:
    print(child['name'])
    # print the names of the children of the second layer
    for subchild in child['children']:
        print('  ' + subchild['name'])
        # print the names of the children of the third layer
        if 'children' in subchild:
            for subsubchild in subchild['children']:
                print('    ' + subsubchild['name'])
```
